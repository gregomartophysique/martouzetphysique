d’un interféromètre optique de type Michelson. La résolution spectrale permet de faciliter 
considérablement la recherche du contact optique en lumière blanche et offre une grande 
précision de réglage de l’orientation des miroirs. En effet, les interférences introduites par l’interféromètre peuvent être interprétées comme une fonction de transfert optique appliquée au spectre 
incident. La transformée de Fourier du spectre mesuré en sortie d’interféromètre permet d’extraire 
en une fraction de seconde la différence de chemin optique entre les deux bras de l’interféromètre 
ainsi qu’un critère sur la qualité du parallélisme des miroirs.
