# Suite de Fibonacci
### Approche naïve

```python
def fibonacci(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibonacci(n-1)+fibonacci(n-2)
    
fibonacci( 10)
```

### Calcul de bas en haut 

```python
def fibonacci_bas_en_haut(n):
    if n==0:
        return 0
    a, b = 0,1
    for i in range(n-1):
        a, b = b, a+b
    return b

fibonacci_bas_en_haut(1000)
```

### Calul de haut en bas avec mémoïsation

```python
d = {0:0,1:1}
def fibonacci_memoisation(n):
    if n in d:
        return d[n]
    d[n] = fibonacci_memoisation(n-1)+fibonacci_memoisation(n-2)
    return d[n]

fibonacci_memoisation(992)
```

### Affichage de la solution
Modifier l'algorithme "bas en haut" pour que qu'il ne donne pas que le nombre de pièce minimal mais aussi la solution associée.
La fonction devra alors renvoyer un tuple contenant la solution et une liste avec toutes les pièces utilisées (éventuellement en plusieurs exemplaires).

```python
def rendu_solution( pieces, somme ):
    pieces.sort()
    # initialisation
    d = {0:(0,[])}
    # relation de récurence
    for s in range(1, somme+1):
        minimum = d[s-pieces[0]][0]+1
        liste_optimale = d[s-pieces[0]][1][:]
        piece_a_garder = pieces[0]
        for p in pieces[1:]:
            if s-p>=0:
                if d[s-p][0]+1<minimum:
                    piece_a_garder = p
                    liste_optimale = d[s-p][1][:]
                    minimum = d[s-p][0]+1
        liste_optimale.append( piece_a_garder )   
        d[s] = (minimum, liste_optimale[:])
    return d[somme]
    
print( rendu_solution([1,2,5], 12) )
print( rendu_solution([1,3,4], 6) )
print( rendu_solution([1,2,5], 101) )
```
